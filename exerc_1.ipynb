```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio, display
import time
```

```python
t=0.01
fs = 44100
ts=t/fs
t=np.arange(0, t, ts)

freqs=[500, 5000, 10000]

# Code to play each frequency
print("Playing each frequency:")
for freq in freqs:
    print(f"Playing {freq} Hz")
    wave = np.sin(2 * np.pi * freq * t)
    display(Audio(wave, rate=fs))
    time.sleep(2) # Pause for 2 seconds between sounds
```

```python
plt.plot(t, np.cos(2*np.pi*freqs[0]*t))
```
