## Load Audio File

### Subtask:
Load the 'handel.wav' file using a suitable library (e.g., `scipy.io.wavfile` or `librosa`) to extract the audio data and its original sampling rate.


**Reasoning**:
The subtask requires loading the 'handel.wav' file. I will use `scipy.io.wavfile` to read the audio data and its sampling rate, then print these details for verification.



```python
from scipy.io import wavfile

# Load the 'handel.wav' file
original_sampling_rate, audio_data = wavfile.read('/content/handel.wav')

# Print the original sampling rate and the shape of the audio data
print(f"Original Sampling Rate: {original_sampling_rate} Hz")
print(f"Audio Data Shape: {audio_data.shape}")
```

## Plot Time Domain Signal

### Subtask:
Generate and display a plot of the loaded audio signal in the time domain. The x-axis should represent time, and the y-axis should represent the amplitude of the signal. Ensure the plot has appropriate labels and a title.


**Reasoning**:
To visualize the audio signal in the time domain, I need to plot the audio data against time. This involves importing necessary libraries like `matplotlib.pyplot` and `numpy`, creating a time array based on the sampling rate, and then generating the plot with proper labels and a title.



```python
import matplotlib.pyplot as plt
import numpy as np

# Create a time array for the x-axis
time = np.linspace(0, len(audio_data) / original_sampling_rate, num=len(audio_data))

# Plot the audio signal in the time domain
plt.figure(figsize=(12, 6))
plt.plot(time, audio_data)
plt.title('Audio Signal in Time Domain')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

## Play Audio at Original Sampling Rate

Write and execute Python code to play the 'handel.wav' audio signal at double its original sampling frequency. Listen carefully to how the sound changes compared to the original.

## Analyze Frequency Spectrum at Original Sampling Rate

### Subtask:
Calculate and plot the frequency spectrum of the `audio_data` when played at its `original_sampling_rate`. Ensure the x-axis represents frequency (Hz) and the y-axis represents amplitude (or magnitude). The plot should have appropriate labels and a title.

**Reasoning**:
To analyze the frequency spectrum, I need to compute the Fast Fourier Transform (FFT) of the audio data. Then, I will generate the corresponding frequency axis and plot the magnitude of the FFT against these frequencies. This requires `numpy` for FFT and `matplotlib.pyplot` for plotting.



```python
from scipy.fft import fft, fftfreq

# Perform FFT on the audio data
N = len(audio_data)  # Number of samples
T = 1.0 / original_sampling_rate  # Sample spacing (inverse of sampling rate)

yf = fft(audio_data)
xf = fftfreq(N, T)[:N//2] # Frequencies for the positive half of the spectrum

# Plot the frequency spectrum (magnitude)
plt.figure(figsize=(12, 6))
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.title('Frequency Spectrum at Original Sampling Rate')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

## Analyze Frequency Spectrum at Double Sampling Rate

### Subtask:
Calculate and plot the frequency spectrum of the `audio_data` when played at `double_sampling_rate`. Ensure the x-axis represents frequency (Hz) and the y-axis represents amplitude (or magnitude). The plot should have appropriate labels and a title.

**Reasoning**:
To analyze the frequency spectrum at double the sampling rate, I will perform the Fast Fourier Transform (FFT) on the audio data using the `double_sampling_rate`. Then, I will compute the corresponding frequencies and plot the magnitude of the FFT against these frequencies, similar to the original sampling rate analysis.



```python
from scipy.fft import fft, fftfreq

# Perform FFT on the audio data
N = len(audio_data)  # Number of samples
T_double = 1.0 / double_sampling_rate  # Sample spacing for double sampling rate

yf_double = fft(audio_data)
xf_double = fftfreq(N, T_double)[:N//2] # Frequencies for the positive half of the spectrum

# Plot the frequency spectrum (magnitude)
plt.figure(figsize=(12, 6))
plt.plot(xf_double, 2.0/N * np.abs(yf_double[0:N//2]))
plt.title('Frequency Spectrum at Double Sampling Rate')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

Then, I will compute the corresponding frequencies and plot the magnitude of the FFT against these frequencies, similar to the previous analyses.



```python
from scipy.fft import fft, fftfreq

# Perform FFT on the audio data
N = len(audio_data)  # Number of samples
T_quadruple = 1.0 / quadruple_sampling_rate  # Sample spacing for quadruple sampling rate

yf_quadruple = fft(audio_data)
xf_quadruple = fftfreq(N, T_quadruple)[:N//2] # Frequencies for the positive half of the spectrum

# Plot the frequency spectrum (magnitude)
plt.figure(figsize=(12, 6))
plt.plot(xf_quadruple, 2.0/N * np.abs(yf_quadruple[0:N//2]))
plt.title('Frequency Spectrum at Quadruple Sampling Rate')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

## Summary of Observations and Relationship between Playback and Frequency Spectra

### Observations:
When the audio is played at:
- **Original Sampling Rate (8192 Hz)**: The audio plays at its intended speed and pitch.
- **Double Sampling Rate (16384 Hz)**: The audio plays twice as fast and sounds higher in pitch. This is because the same number of samples are being played back in half the time.
- **Quadruple Sampling Rate (32768 Hz)**: The audio plays four times as fast and sounds even higher in pitch, a further acceleration compared to the original.

### Relationship with Frequency Spectra:
The frequency spectra (obtained by FFT of the `audio_data` but interpreted with different sampling rates for the frequency axis `xf`) reveal a crucial aspect:

- The underlying **spectral content** (the actual frequencies present in the `audio_data` itself) does not change. The FFT operation on `audio_data` will always produce the same `yf` (frequency domain representation) regardless of the intended playback sampling rate, as `yf` is derived directly from `audio_data`.

- What *does* change is the **interpretation of the frequency axis (xf)**. The `fftfreq` function generates frequencies based on the number of samples `N` and the sampling interval `T` (which is `1/sampling_rate`).
    - At the `original_sampling_rate`, the highest frequency represented (Nyquist frequency) is `original_sampling_rate / 2`.
    - At `double_sampling_rate`, the `xf_double` axis extends to `double_sampling_rate / 2`. The spectral peaks appear at proportionally higher frequencies on this new scale, but the actual frequency components within the signal itself haven't shifted. It means that the *range* of frequencies that can be represented (and thus heard) has expanded.
    - At `quadruple_sampling_rate`, the `xf_quadruple` axis extends to `quadruple_sampling_rate / 2`. Again, the spectral content is stretched across a wider frequency range, indicating a higher maximum representable frequency.

**In essence, when you increase the sampling rate for playback without changing the underlying audio data, you are telling the playback device to 'stretch' the original sound data over a wider perceived frequency range, making the sound play faster and at a higher pitch. The FFT plot correctly reflects this by scaling the frequency axis according to the sampling rate used for its calculation.**
