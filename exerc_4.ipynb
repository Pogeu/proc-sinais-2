### Procedimento para Cálculo da Resposta ao Impulso de uma Sala

A Resposta ao Impulso (IR) define completamente as propriedades acústicas lineares e invariantes no tempo de um recinto. O procedimento padrão envolve:

1.  **Escolha do Método de Excitação:**
    *   **Método Direto (Impulsivo):** Uso de uma fonte que se aproxima de um Delta de Dirac, como um tiro de espoleta ou estouro de balão. É simples, mas sofre com baixa relação sinal-ruído (SNR).
    *   **Sine Sweep (Varredura de Senoide):** O método mais preciso. Emite-se um sinal senoidal cuja frequência varia logaritmicamente. A IR é obtida através da deconvolveção do sinal gravado com o sinal original.
    *   **MLS (Maximum Length Sequence):** Uso de ruído pseudo-aleatório. Requer que o sistema seja perfeitamente invariante no tempo.

2.  **Configuração de Equipamentos:**
    *   Posiciona-se uma fonte omnidirecional (como um dodecaedro) no local da fonte sonora.
    *   Posiciona-se microfones de medição (geralmente omnidirecionais) nos pontos de recepção desejados.

3.  **Gravação e Processamento:**
    *   Grava-se a resposta da sala.
    *   No caso do Sine Sweep, aplica-se um filtro inverso para extrair a IR, separando as distorções harmônicas da resposta linear.

### Limitações e Considerações

*   **Ruído de Fundo:** O ruído ambiental pode mascarar o decaimento da cauda da reverberação, limitando o alcance dinâmico da medição.
*   **Não-Linearidades:** Equipamentos (alto-falantes e microfones) introduzem coloração. É necessário compensar a resposta do próprio sistema de medição.
*   **Variância Temporal:** Mudanças de temperatura ou movimento de pessoas durante a medição invalidam métodos como o MLS.
*   **Posicionamento:** A IR é específica para um par fonte-receptor. Mover o microfone alguns centímetros altera os padrões de interferência em altas frequências.
